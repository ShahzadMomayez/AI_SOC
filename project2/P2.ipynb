# Linear Regression

Main form of simple linear regression function: 
$$f(x) = \alpha x + \beta$$

The model gets the best regression fit line by finding the best $\alpha$ and $\beta$ values. 

$\alpha$ : intercept

$\beta$: slope

$$ \hat{\alpha} = \frac{\Sigma{(x_i - \hat{x})(y_i - \hat{y})}}{\Sigma{(x_i - \hat{x})^2}}
$$ 
$$ \hat{\beta} = \hat{y} - \hat{\alpha} \hat{x}$$

In linear regression, the objective is to determine the optimal values of $\alpha$ and $\beta$ that accurately represent the data. To achieve this, the method of least squares is employed, which entails minimizing the sum of the squared deviations between the predicted and actual values of the dependent variable.



```python
def linear_regression(input, output):
    sum_input = np.sum(input)
    sum_output = np.sum(output)
    # Calculate the sum of the input_feature and output.
    # Calculate the sum of the product of the output and the input_feature.
    # Calculate the sum of the squared values of the input_feature.
    # Apply the formula to calculate the slope.
    # Apply the formula to calculate the intercept.
    mult_in_out = np.sum(input * output)
    SSi = np.sum(input ** 2)
    
    intercept = (mult_in_out - (len(input) * (sum_input / len(input)) * (sum_output / len(output)))) / (SSi - len(input) * ((sum_input / len(input)) ** 2))
    slope = (sum_output / len(output)) - intercept * (sum_input / len(input))
    print (intercept, slope)
    return (intercept, slope)
```

```python
#Now complete this function to predict the value of given data based on the calculated intercept and slope
def get_regression_predictions(input, intercept, slope):
    # TO DO:
    # calculate the predicted values:
    predicted_values = [(intercept * x) + slope for x in input]
    return (predicted_values)
```

```python
intercept, slope = linear_regression(X_train['curbweight'], y_train['price'])
y_pred = get_regression_predictions(X_test['curbweight'], intercept, slope)
```
